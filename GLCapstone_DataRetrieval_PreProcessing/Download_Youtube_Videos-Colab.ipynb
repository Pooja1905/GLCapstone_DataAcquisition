```python
# Read the video from specified path 
def video_to_frames_converter(input_loc, output_loc, imageNameHead):
    cam = cv2.VideoCapture(input_loc) 
  
    try: 
      
        # creating a folder named data 
        if not os.path.exists(output_loc): 
            os.makedirs(output_loc) 
  
    # if not created then raise error 
    except OSError: 
        print ('Error: Creating directory of data') 
  
    # frame 
    currentframe = 0
  
    while(True): 
      
        # reading from frame 
        ret,frame = cam.read() 
  
        if ret: 
            # if video is still left continue creating images 
            name = output_loc + '/'+ imageNameHead + "_%#05d.jpg" % (currentframe+1)         
            #print(name)
            # writing the extracted images 
            cv2.imwrite(name, frame) 
  
            # increasing counter so that it will 
            # show how many frames are created 
            currentframe += 1
        else: 
            break
  
    # Release all space and windows once done 
    cam.release() 
    cv2.destroyAllWindows() 
```

```python
list_of_videos = os.listdir('Clipped_Videos/')
input_loc = 'Clipped_Videos/'
output_loc = 'Decomposed_Videos_To_Images'
#list_of_videos.remove('.ipynb_checkpoints')
i = 0
sans_C = []
for clipped_vid in list_of_videos:
    print(clipped_vid)
    sans_mp4 = clipped_vid.split('.')[0]
    sans_C.append(sans_mp4.split('_'))    
    image_target_name = sans_C[i][1]
    video_to_frames_converter(input_loc+clipped_vid, output_loc, image_target_name)  
    i = i + 1   
```
